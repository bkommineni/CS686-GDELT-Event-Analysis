# GDELT Event Data Analysis

```python
from pyspark.sql.types import StructType, StructField, FloatType, LongType, StringType, IntegerType
from pyspark.shell import spark
```

```python
attrs = []
f = open('./GDELT-EVENTS-ATTRIBUTES.txt')

for line in f:
    tokens = line.split(',')
    if tokens[1].strip() == "INTEGER":
        attrs.append(StructField(tokens[0].strip(), IntegerType(), True))
    elif tokens[1].strip() == "STRING":
        attrs.append(StructField(tokens[0].strip(), StringType(), True))
    elif tokens[1].strip() == "FLOAT":
        attrs.append(StructField(tokens[0].strip(), FloatType(), True))
    
schema = StructType(attrs)
schema
```

```python
df_data_2015 = spark.read.format('CSV').option('sep', '\t').schema(schema).load('/Users/npbandal/BigData/project3/p3-d2-girlswhocode/DataNewZip2015/2015*.export.CSV')
df_data_2015.take(1)
```

```python
df_data_2015.createOrReplaceTempView("POPEVENT2015")

popevent = spark.sql("SELECT Actor1Geo_FullName,EventRootCode,DATEADDED,Actor1Geo_CountryCode,Actor1Geo_FullName,NumMentions FROM POPEVENT2015 where NumMentions in (SELECT MAX(NumMentions) FROM POPEVENT2015)").collect()
print('Most viewed event/news:', popevent)

```
